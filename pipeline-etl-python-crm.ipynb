```python
import pandas as pd

# Seus dados atualizados
data = {
    'ID': [1, 2, 3, 4, 5],
    'Nome': ['Vânia Bordin', 'Marcos Silva', 'Ana Paula', 'Roberto Junior', 'Carla Souza'],
    'Score': [95, 40, 75, 20, 88],
    'Setor': ['Data Science', 'BI', 'Engenharia', 'Marketing', 'Data Science'],
    'Ultima_Compra': ['2024-05-01', '2024-04-15', '2024-05-10', '2024-03-20', '2024-05-12']
}

df = pd.DataFrame(data)

# Salvando no diretório do Colab sem a coluna de índice extra
df.to_csv('/content/clientes_desafio.csv', index=False)

print("Base de dados 'clientes_desafio.csv' criada com sucesso em /content/!")
```

**1. Extração (Extract)**


Ler os dados do arquivo .csv

```python

def extract():
    # Lendo o arquivo criado no diretório do Colab
    return pd.read_csv('/content/clientes_desafio.csv')
```

2. Transformação (Transform)


```python
def transform(df):
    """
    Simulação de lógica de IA:
    Classifica o cliente pelo Score e gera uma recomendação por Setor.
    """

    def gerar_insight(row):
        # Lógica de Classificação (IA de Negócio)
        if row['Score'] >= 80:
            status = "Premium"
            mensagem = f"Parabéns {row['Nome']}! Como especialista em {row['Setor']}, você tem acesso a mentorias VIP."
        elif row['Score'] >= 50:
            status = "Ativo"
            mensagem = f"Olá {row['Nome']}, confira os novos workshops de {row['Setor']} disponíveis."
        else:
            status = "Alerta de Churn"
            mensagem = f"Sentimos sua falta, {row['Nome']}. Preparamos um cupom especial para retomar seus estudos em {row['Setor']}."

        return pd.Series([status, mensagem])

    # Aplicando a transformação e criando duas novas colunas
    df[['Status', 'Mensagem_IA']] = df.apply(gerar_insight, axis=1)
    return df
```

**3. Carregamento (Load)**

```python
def load(df):
    # Salvando o resultado final em um novo arquivo CSV
    output_path = '/content/clientes_com_insights.csv'
    df.to_csv(output_path, index=False)
    return output_path
```

** EXECUÇÃO DO PIPELINE (O Coração do ETL)**

```python
print("Iniciando o Pipeline ETL...")

# Passo 1: Extrair
df_raw = extract()
print(f"Extração concluída: {len(df_raw)} registros encontrados.")

# Passo 2: Transformar
df_processed = transform(df_raw)
print("Transformação inteligente concluída com sucesso.")

# Passo 3: Carregar
caminho_final = load(df_processed)
print(f"Pipeline finalizado! O arquivo foi salvo em: {caminho_final}")

# Visualizando o resultado final aqui no Colab
df_processed.head()
```
